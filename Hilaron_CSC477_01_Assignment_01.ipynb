```python
# loading the dataset
import pandas as pd

'''Task 1

Loading and Basic Information about the Dataset
Display the first 10 rows of the dataset
Show basic information bout the size number of rows and columns
List column name and their types.'''

# load dataset using pandas
path = '/content/Life Expectancy Data.csv'
df = pd.read_csv(path)

first_10 = df.head(10)  # show first 10 rows
rows, cols = df.shape  # show number of rows and columns
cols_name_dtype = df.dtypes  # show the columns and the data types

#  show the results
task1_result = f"\nDisplay the first 10 rows of the dataset:\n\n{first_10}\n\
\nnumber of rows: {rows} and number of columns: {cols}\n\
\nList the column names and their data types:\n\n{cols_name_dtype}"


print(task1_result)
```

```python

''' Task 2

Summary Statistics for Numeric Columns
in this task, identify all numeric columns i display their name.
for each numeric column the mean, median standard deviation, variance,
minimum value, maximum value, 25th percentile (Q1), 75th percentile (Q3),
skewness and the kurtosis are computed.'''

# Identify all numeric columns in the dataset and display their names.
numeric_cols = [c for c in df.columns if df[c].dtype == 'int64' or df[c].dtype == 'float64']

# For each numeric column, compute and display Mean Median std
summary_statistics = pd.DataFrame()

for col in numeric_cols:
    summary_statistics.loc[col, 'Mean'] = df[col].mean()
    summary_statistics.loc[col, 'Median'] = df[col].median()
    summary_statistics.loc[col, 'Standard Deviation'] = df[col].std()
    summary_statistics.loc[col, 'Variance'] = df[col].var()
    summary_statistics.loc[col, 'Minimum Value'] = df[col].min()
    summary_statistics.loc[col, 'Maximum Value'] = df[col].max()
    summary_statistics.loc[col,
                    '25th percentile (1st quartile)'] = df[col].quantile(.25)
    summary_statistics.loc[col,
                    '75th percentile (3rd quartile)'] = df[col].quantile(.75)
    summary_statistics.loc[col, "Skewness"] = df[col].skew()
    summary_statistics.loc[col, 'Kurtosis'] = df[col].kurtosis()

# results
summary_statistics
```

```python
'''Task 3

Interpretation of Statistics for Information Discovery
'''
# Information discovery I

schooling_mean = df['Schooling'].mean()  # 11.992792792792793
schooling_std = df['Schooling'].std()  # 3.3589197211023563
schooling_kurtosis = df['Schooling'].kurtosis()  # 0.886151268906703

print("__Information discovery I results__\n")

print(f"schooling_mean: {schooling_mean}\nschooling_std: {schooling_std}\n\
schooling_kurtosis: {schooling_kurtosis}")

# Mean 11.99 years, Standard Deviation 3.36 and Kurtosis 0.89.
# This suggest that most countries have aroud 12 years of schooling,
# but few countries deviate from this, with some having either significantly
# shorter or longer periods of education.
# The moderately high kurtosis suggest the presence of a few extreme value
# either low or high schooling years.

# Information discovery II

population_mean = df['Population'].mean()  # 12753375.120052494
population_std =  df['Population'].std()  # 61012096.508427925
population_skewness =  df['Population'].skew()  # 15.916235598428203

print("\n__Information discovery II results__\n")

print(f"population_mean: {population_mean}\npopulation_std: {population_std}\n\
population_skewness: {population_skewness}")

# Mean 12,753,375 people, Standard Deviation 61,012,096 people, Skewness (15.92)
# The population data is heavily influenced by a few outliers with extremely
# large populations. Most countries have populations much higher than the mean
# of 12.75 trillion. the right skewed distribution suggest that the majority
# of countries are small or medium-sized in terms of population, but due to some
# very populated countries the mean is pull upward, leading to high skewness.

# Information discovery III

hiv_aids_median = df[' HIV/AIDS'].median()  # 0.1
hiv_aids_variance = df[' HIV/AIDS'].var()  # 25.783895744141844
hiv_aids_kurtosis = df[' HIV/AIDS'].kurtosis()  # 34.892007872835045

print("\n__Information discovery III results__\n")

print(f"hiv_aids_median: {hiv_aids_median}\nhiv_aids_variance:\
{hiv_aids_variance}\n\
# hiv_aids_kurtosis: {hiv_aids_kurtosis}")

# HIV/AIDS distribution is highly skewed:
# while most countries have very low HIV/AIDS death rates indicate by low median
# a few country experience much higher rates, as shown by the high variance
# and kurtosis. These outliers contribute to the high kutosis and suggest
# that HIV/AIDS is still a severe issue in some parts of the world
```
